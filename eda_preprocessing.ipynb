# 2. DATA PRE-PROCESSING AND FEATURE ENGINEERING

Bu bölümde EDA bulgularına dayanarak kapsamlı veri ön işleme ve özellik mühendisliği gerçekleştireceğiz:

1. **Eksik Değer İşleme**: IterativeImputer (sayısal değişkenler için) + “Missing” kategorisi ekleme (kategorik değişkenler için)
2. **Kategorik Değişken Kodlama**: OrdinalEncoder (düşük kardinaliteli değişkenler) + Frequency Encoding (yüksek kardinaliteli değişkenler)
3. **Özellik Mühendisliği**: Yaş grupları, kan grubu özellikleri, vs.
4. **Sayısal Değişken Normalizasyonu**: RobustScaler ile ölçeklendirme
5. **Pipeline Seviyesi Kod**: Tüm işlemler Pipeline ve ColumnTransformer ile modüler ve yeniden kullanılabilir hale getirildi
6. **Korelasyon Analizi**: İşlem sonrası hedef değişken ile özellikler arasındaki korelasyonlar hesaplandı ve görselleştirildi


```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.experimental import enable_iterative_imputer 
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import OrdinalEncoder, RobustScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
warnings.filterwarnings('ignore')
```

```python

df_original = df.copy()
df_processed = df.copy()
print(f"Orijinal veri seti boyutu : {df_original.shape[0]:,} satır × {df_original.shape[1]} sütun")
print(f"İşlenecek veri seti boyutu: {df_processed.shape[0]:,} satır × {df_processed.shape[1]} sütun")
target_column = "TedaviSuresi_numeric"
print(f"Hedef değişken: {target_column}")
feature_columns = [col for col in df_processed.columns if col not in ["HastaNo", target_column]]
print(f"Toplam özellik sütunu: {len(feature_columns)}")
for i, col in enumerate(feature_columns, 1):
    print(f"{i:2d}. {col}")
print(f"Toplam satır sayısı     : {df_processed.shape[0]:,}")
print(f"Toplam sütun sayısı     : {df_processed.shape[1]}")
print(f"Özellik sütunu sayısı   : {len(feature_columns)}")
```

```python

missing_analysis = pd.DataFrame({
    "Sütun": feature_columns,
    "Eksik_Sayısı": df_processed[feature_columns].isnull().sum().values,
    "Eksik_Yüzdesi": (df_processed[feature_columns].isnull().sum().values / len(df_processed)) * 100,
    "Veri_Tipi": df_processed[feature_columns].dtypes.values,
    "Benzersiz_Sayısı": df_processed[feature_columns].nunique().values
})

def determine_imputation_strategy(row, low=5, mid=20):
    if row["Eksik_Yüzdesi"] == 0:
        return "İmputasyon gerekmez"
    elif row["Veri_Tipi"] in ["int64", "float64"]:
        # Sayısal değişkenler
        if row["Eksik_Yüzdesi"] < low:
            return "Median imputasyon"
        else:
            return "Iterative Imputer (MICE)"
    else:
        # Kategorik değişkenler
        if row["Eksik_Yüzdesi"] < mid:
            return "'Missing' kategorisi ekle"
        else:
            return "'Missing' kategorisi ekle veya kategori birleştir"

missing_analysis["İmputasyon_Stratejisi"] = missing_analysis.apply(determine_imputation_strategy, axis=1)

display(missing_analysis.sort_values("Eksik_Yüzdesi", ascending=False))

plt.figure(figsize=(12, 7))
missing_analysis_sorted = missing_analysis.sort_values("Eksik_Yüzdesi", ascending=False)

colors = [
    "green" if x < 5 else "orange" if x < 20 else "red" 
    for x in missing_analysis_sorted["Eksik_Yüzdesi"]
]

plt.bar(
    missing_analysis_sorted["Sütun"],
    missing_analysis_sorted["Eksik_Yüzdesi"],
    color=colors,
    alpha=0.8
)
plt.title("Sütunlara Göre Eksik Değer Yüzdeleri", fontsize=14, fontweight="bold")
plt.xlabel("Sütunlar", fontsize=12)
plt.ylabel("Eksik Değer Yüzdesi (%)", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", alpha=0.3)

from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor="green", label="Düşük (<5%)"),
    Patch(facecolor="orange", label="Orta (5-20%)"),
    Patch(facecolor="red", label="Yüksek (>20%)")
]
plt.legend(handles=legend_elements, loc="upper right")
plt.tight_layout()
plt.show()
print(f"Toplam eksik değer sayısı : {df_processed[feature_columns].isnull().sum().sum():,}")
print(f"Eksik değer içeren sütun sayısı : {(missing_analysis['Eksik_Sayısı'] > 0).sum()}")
print(f"Tüm sütun sayısı : {len(feature_columns)}")
print(f"En çok eksik değerli sütun : {missing_analysis.loc[missing_analysis['Eksik_Sayısı'].idxmax(), 'Sütun']}")
```

```python

updated_feature_columns = [col for col in df_processed.columns if col not in ['HastaNo', target_column]]

categorical_columns = df_processed[updated_feature_columns].select_dtypes(include=['object']).columns.tolist()
numerical_columns = df_processed[updated_feature_columns].select_dtypes(include=[np.number]).columns.tolist()

print(f"Kategorik sütunlar ({len(categorical_columns)} adet):")
for i, col in enumerate(categorical_columns, 1):
    unique_count = df_processed[col].nunique()
    missing_count = df_processed[col].isnull().sum()
    print(f"  {i:2d}. {col} (Benzersiz: {unique_count}, Eksik: {missing_count})")

print(f"\nSayısal sütunlar ({len(numerical_columns)} adet):")
for i, col in enumerate(numerical_columns, 1):
    missing_count = df_processed[col].isnull().sum()
    print(f"  {i:2d}. {col} (Eksik: {missing_count})")

def determine_encoding_strategy(column, threshold=10):
    unique_count = df_processed[column].nunique()
    if unique_count <= threshold:
        return 'OrdinalEncoder (Low Cardinality)'
    else:
        return 'Frequency Encoding (High Cardinality)'

encoding_strategies = {col: determine_encoding_strategy(col) for col in categorical_columns}

print(f"\nKodlama stratejileri:")
for col, strategy in encoding_strategies.items():
    print(f"  {col}: {strategy}")

high_cardinality_cols = [col for col, strategy in encoding_strategies.items() 
                         if 'High Cardinality' in strategy]
print(f"\nYüksek kardinaliteli sütunlar (Frequency Encoding): {high_cardinality_cols}")

for col in high_cardinality_cols:
    df_processed[col] = df_processed[col].fillna('Missing')
    freq_map = df_processed[col].value_counts().to_dict()
    df_processed[col + '_freq_encoded'] = df_processed[col].map(freq_map)
    print(f"  {col} -> {col}_freq_encoded eklendi (Frequency encoding)")

ordinal_cols = [col for col, strategy in encoding_strategies.items() 
                if 'OrdinalEncoder' in strategy]
print(f"\nDüşük kardinaliteli sütunlar (OrdinalEncoder uygulanacak): {ordinal_cols}")

for col in ordinal_cols:
    df_processed[col] = df_processed[col].fillna('Missing')
```

```python
from sklearn.experimental import enable_iterative_imputer  
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import OrdinalEncoder, RobustScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import pandas as pd
import numpy as np

final_feature_columns = [col for col in df_processed.columns if col not in ["HastaNo", target_column]]

final_categorical_columns = df_processed[final_feature_columns].select_dtypes(include=["object"]).columns.tolist()
final_numerical_columns = df_processed[final_feature_columns].select_dtypes(include=[np.number]).columns.tolist()

print(f"Final kategorik sütunlar ({len(final_categorical_columns)} adet): {final_categorical_columns}")
print(f"Final sayısal sütunlar ({len(final_numerical_columns)} adet): {final_numerical_columns}")

ordinal_cols = ["Cinsiyet", "KanGrubu", "Uyruk", "Bolum", "UygulamaSuresi"]

freq_encoded_cols = [
    "KronikHastalik_freq_encoded",
    "Alerji_freq_encoded",
    "Tanilar_freq_encoded",
    "TedaviAdi_freq_encoded",
    "TedaviSuresi_freq_encoded",
    "UygulamaYerleri_freq_encoded"
]

all_numeric_cols = pd.Index(final_numerical_columns + freq_encoded_cols).drop_duplicates().tolist()

categorical_pipeline = Pipeline([
    ("encoder", OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1))
])

numerical_pipeline = Pipeline([
    ("imputer", IterativeImputer(random_state=42)),
    ("scaler", RobustScaler())
])

preprocessor = ColumnTransformer([
    ("categorical", categorical_pipeline, ordinal_cols),
    ("numerical", numerical_pipeline, all_numeric_cols)
], remainder="drop")  # kalanları alma

print(f"  - Sayısal pipeline: IterativeImputer + RobustScaler")
print(f"  - Kategorik pipeline: OrdinalEncoder")
print(f"  - High-cardinality kolonlar: Frequency Encoding ile sayısallaştırıldı")

X = df_processed[ordinal_cols + all_numeric_cols]
y = df_processed[target_column]

X_transformed = preprocessor.fit_transform(X)

print(f"  - Orijinal özellik sayısı: {X.shape[1]}")
print(f"  - Dönüştürülmüş özellik sayısı: {X_transformed.shape[1]}")
print(f"  - Satır sayısı: {X_transformed.shape[0]}")

feature_names = preprocessor.get_feature_names_out()
feature_names = [name.replace("categorical__", "").replace("numerical__", "") for name in feature_names]

X_transformed_df = pd.DataFrame(X_transformed, columns=feature_names, index=X.index)

print(f"\n📊 Dönüştürülmüş veri seti oluşturuldu:")
print(f"  - Sütun sayısı: {X_transformed_df.shape[1]}")
print(f"  - Satır sayısı: {X_transformed_df.shape[0]}")
print(f"  - Eksik değer sayısı: {X_transformed_df.isnull().sum().sum()}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline


final_feature_columns = [col for col in df_processed.columns if col not in ['HastaNo', target_column]]

final_categorical_columns = df_processed[final_feature_columns].select_dtypes(include=['object']).columns.tolist()
final_numerical_columns = df_processed[final_feature_columns].select_dtypes(include=[np.number]).columns.tolist()

print(f"Final kategorik sütunlar ({len(final_categorical_columns)} adet): {final_categorical_columns}")
print(f"Final sayısal sütunlar ({len(final_numerical_columns)} adet): {final_numerical_columns}")

categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore'))
])

numerical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

preprocessor = ColumnTransformer([
    ('categorical', categorical_pipeline, final_categorical_columns),
    ('numerical', numerical_pipeline, final_numerical_columns)
], remainder='drop')

X = df_processed[final_feature_columns]
y = df_processed[target_column]

X_transformed = preprocessor.fit_transform(X)
feature_names = preprocessor.get_feature_names_out()

X_transformed_df = pd.DataFrame(
    X_transformed, columns=feature_names, index=X.index, dtype=float
)

X_transformed_df_with_target = X_transformed_df.copy()
X_transformed_df_with_target[target_column] = pd.to_numeric(y, errors='coerce')

print(f"\nKorelasyon analizi için veri seti hazırlandı:")
print(f"  - Özellik sayısı: {X_transformed_df.shape[1]}")
print(f"  - Hedef değişken: {target_column}")
print(f"  - Toplam sütun: {X_transformed_df_with_target.shape[1]}")

numeric_columns = X_transformed_df_with_target.select_dtypes(include=[np.number]).columns.tolist()
correlation_matrix = X_transformed_df_with_target[numeric_columns].corr()

if target_column in numeric_columns:
    target_correlations = correlation_matrix[target_column].drop(target_column).sort_values(key=abs, ascending=False)
else:
    print(f" Hedef değişken {target_column} sayısal sütunlarda bulunamadı!")
    target_correlations = pd.Series(dtype=float)

print(f"\nHedef değişken ile en yüksek korelasyonlu özellikler (Top 20):")
print("="*60)
if not target_correlations.empty:
    for i, (feature, corr) in enumerate(target_correlations.head(20).items(), 1):
        direction = "Pozitif" if corr > 0 else "Negatif"
        strength = "Güçlü" if abs(corr) > 0.3 else "Orta" if abs(corr) > 0.1 else "Zayıf"
        print(f"{i:2d}. {feature}: {corr:6.3f} ({direction}, {strength})")
else:
    print("Hiç korelasyon bulunamadı!")
```
